## 訓練LSTM模型

```python
from keras.models import Sequential
from keras.layers import LSTM, Dense
from sklearn.metrics import mean_squared_error
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('./df_merged.csv')
display(df)
```

```python
df.columns
```

```python
X = df.drop('單價元平方公尺', axis=1)  # 刪除 '單價元平方公尺' 欄位，其餘欄位為目標變數
y = df['單價元平方公尺']
display(X)
```

```python
X_train, X_test_val, y_train, y_test_val = train_test_split(X, y, test_size=0.2, random_state=23)
X_test, X_val, y_test, y_val = train_test_split(X_test_val, y_test_val, test_size=0.5, random_state=23)
```

```python
# 定義模型
model = Sequential()
model.add(LSTM(units=64, input_shape=(X_train.shape[1], 1)))  # 設定LSTM的輸入形狀，timesteps是時間步數，input_dim是輸入特徵的維度
model.add(Dense(1))  # 輸出層只有一個節點，對應目標變數的預測值

# 編譯模型
model.compile(loss='mean_squared_error', optimizer='adam')

# 訓練模型
model.fit(X_train, y_train, epochs=10, batch_size=32)

# 使用模型進行預測
y_pred_val = model.predict(X_val)
```

```python
# 計算MSE
mse = mean_squared_error(y_val, y_pred_val)
print("mse:", mse)

# 計算RMSE
rmse = mean_squared_error(y_val, y_pred_val, squared=False)
print("rmse:", rmse)
```

```python
# 使用模型進行預測
y_pred = model.predict(X_test)
```

```python
# 計算MSE
mse = mean_squared_error(y_test, y_pred)
print("mse:", mse)

# 計算RMSE
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("rmse:", rmse)
```

```python
model.save('LSTM.h5') #儲存模型
```

```python
def changeRoadToNum(input_road):
    if not df_road[df_road['Unnamed: 0'] == input_road].empty:
        return df_road[df_road['Unnamed: 0'] == input_road].iloc[0, 1]
    else:
        return -1

changeRoadToNum('龍泉段一小段')
```
